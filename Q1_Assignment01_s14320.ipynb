```python
def plot_lines(x1,x2,y1,y2):
    #plot triangle and square
    x=np.linspace(-1,1);
    y=np.linspace(-1,1);
    xx=[x1,x2];
    yy=[y1,y2];

    slope_intercept01 = np.polyfit(xx,yy,1)

    #print(slope_intercept01)
    line1=slope_intercept01[0]*x+slope_intercept01[1]
    plt.plot(x,line1)
    plt.axis([-1, 1.0, -1, 1.0])
    print(line1)
    return slope_intercept01
```

```python
line1=plot_lines(-1,0,-1,1)
l1=plotlines(-1,0,-1,1)
line2=plot_lines(1,0,-1,1)
l2=plotlines(1,0,-1,1)
```

```python
def cal_s(N):
  count=0
  for i in range(N):
    xrand=random.uniform(-1,1)
    yrand=random.uniform(-1,1)

    #if ((line1.any()) >= yrand ) & ((line2.any()) >= yrand)  & (-1 <= yrand):
    if (l2[0]*xrand+l2[1] >= yrand ) & (l1[0]*xrand+l1[1] >= yrand)  & (-1 <= yrand):
      plt.plot(xrand,yrand,'bo')
      count+=1
    else:
      plt.plot(xrand,yrand,'go')
  ratio=count/N
  return ratio

no_rad_num=[10,100, 1000, 5000, 10000]
#List to store the s ratio for a 5 times
S_list=[]
#List to store varience for each N
var_list=[]
length = len(no_rad_num)
#5 number of trials for s ratio for list in no_rad_num times
for i in no_rad_num: 
    for j in range(5):
        S=cal_s(i)
        S_list.append(S)
        #Calculating mean and variance of the probabilities for a given N
    Savg=stat.mean(S_list)
    variance=stat.variance(S_list)

            #storing the mean and variance of the probabilities for a given N
    S_list.append(Savg)
    var_list.append(variance)
    print(f'S ratio average for {i} ={Savg}')
    print(f'S ratio variance for {i}={variance}')
    print(f'Probability for {i} = {Savg}')
    print('  ')
```

```python
plt.plot(no_rad_num,var_list,marker='o')
plt.xlabel('Number of iterations')
    #naming the y axis
plt.ylabel('Variance')

    #giving a title to my graph
plt.title('Variance Verse Number of iterations')

    #function to show the plot
plt.show()
```
